```python
'''
#Stack -- ADT
Is a list with the restriction that insertion and deletion 
can only be done from the top.
Last Item In First Item out

#Operations possible on stacks;
-Push(x)
-Pop()
-Top() - Obtain item at the top of the list
-isEmpty() 

Point to note the time complecity of the above operations is constant, hence = O(1)

Stacks can be implemented through
-Arrays
-Linked list

-Array - Stacks 
In array stacks we simply add an element at the end of the stack, to get the last element
we just find the element in the -1 index.
The only downside is that if we exhaust the array we will have to create a new array, 
and hence it will make the time of operation be O(n)

- Linked list - Stacks
For this the best feasible way to access the top of the stack it at the start of the list, this is because
the time taken to access the front of the list is O(1) but if we want to access the end of the list the time is O(n)


# Below is a simple structure of a Stack(Array)
'''   

class Stack():
    def __init__(self, *args, **kwargs):
        self.items = []   ### Stack is a modification of a list

    def push(self, item):
        self.items.append(item)
    
    def pop(self):
        return self.items.pop()
    def get_stack(self):
        return self.items
    def is_empty(self):
        return self.items == []
    def top(self):
        return self.items[-1]
```

```python
#Examples -- Check if Parenthesis are balanced.
# Its check done by IDES
# To implement this we are going to create a stack that will receive the the opening parenthesis i.e ({[,
# While adding the closing parenthisis )}], we will first check if the stack is empty, is so the stack 
#is not balanced, if the stack is not empty, we are to compare the element in the top of the stack to the 
# to the closing parenthesis.

#In this demonstration we start with a string of parenthesis and try to add to the stack one by one
#The goal is to ensure that the stack we create is empty at the end and that the is_balance variable
# is true at the end
#The stack array to be used is the one in the above cell

def is_match(top, paren):
    if top == "(" and paren == ")":
        return True
    elif top == "[" and paren == "]":
        return True
    elif top == "{" and paren == "}":
        return True
    else:
        return False

def is_paren_balanced(paren_string):
    s = Stack() # Initialising a stack item
    is_balanced = True # Start Value is true, we expect it to be true at the end
    index = 0  # helps while looping the paren string

    while index < len(paren_string) and is_balanced:
        
        paren = paren_string[index]
        #Adding the elements - we add if its an oppening paren
        if paren in "([{":
            s.push(paren)
            
        else:
            # When we hit a closing paren
            if s.is_empty():
                is_balanced = False # WE expect the stack to have something
            else:
                #if the stack is not empty, we match the paren at the top of the stack
                #we check if it matches with the closing paren 
                top = s.pop()
                is_balanced = is_match(top, paren)
                # if not is_match(top, paren):
                #     is_balanced = False
       
        
        index += 1 
    return (s.is_empty and is_balanced)

# strings = "()"

is_paren_balanced("()")          
```

```python
# Converting integer values to binary
'''  
To convert an integer into binarry, we divide integer by 2,
and with each reminder we check if the number is divisble by 2
Example
                            This part will be the binary
242 // 2 = 121  and  242%2   == 0
121 // 2 = 60   and  121%2   == 1
60 // 2 = 30    and  60%2    == 0
30 // 2 = 15    and  30%2    == 0
15 // 2 = 7     and  15%2    == 1
7 // 2 = 3      and  7%2     == 1
3 // 2 = 1      and  3%2     == 1
1 // 2 = 0      and  1%2     == 1
                                The binary will be the reverse of this column

To confirm the binary int('1110010', 2)

'''  
# To impelement the above

def convert_binary(number):
    # Initialising a stack item
    s = Stack()
    rem = number
    binary = ''
    #To create the binary inside the stack
    while rem != 0:
        # print(f"{rem=}")
        s.push(rem%2)
        rem = rem //2
    # print(s.get_stack())

    # To convert the stack to string
    while not s.is_empty():
        binary += f"{s.pop()}"

    # binary += "1"  
    # print(binary)

    # print(int(binary, 2))

    return binary if int(binary, 2) == number else "Error"

convert_binary(242)
```

```python
#To reverse a string
#It will be similar to the above method

def reverse_string(string):
    #Creating a stack instance
    s = Stack()
    index = 0
    reversed_string = ""
    while index < len(string):
        item = string[index]
        s.push(item)
        index += 1
    while not s.is_empty():
        reversed_string += s.pop()
    
    return reversed_string

reverse_string("Hello")
```
