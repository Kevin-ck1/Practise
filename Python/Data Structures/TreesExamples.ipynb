```python
#Unique Binary Search Trees
#Given an interger n, return the number of structurally 
# unique BST which has exactly n nodes of unique values 
# form 1 to n.
#      1                    2
#     / \                  / \
#    0   2                1   3
#         \                    \ 
#          3                    4
#           \                       
#            4
def numTrees(n):
    # creating an array of 1s from 0 to n+1
    numTree = [1]* (n+1)
    #Base Case
    # 0 nodes = 1 Tree
    # 1 node = 1 tree

    for nodes in range(2, n+1):
        total = 0
        for root in range(1, nodes + 1):
            left = root - 1
            right = nodes - root
            total += numTree[left] * numTree[right]
        numTree[nodes] = total

    return numTree[n]

numTrees(5)
```

```python
class Node:
    def __init__(self, val):
        self.val = val  
        self.left = None
        self.right = None 

class BST:
    def __init__(self, *args, **kwargs):
        self.root  = None

    def create_bst(self, a):
        node = Node(a.pop(0))
        self.root = node 

        q = [self.root]

        while q:
            cur_node = q.pop(0)
            if len(a) > 0 and a[0] is not None:
                cur_node.left = Node(a[0])
                q.append(cur_node.left)
            if len(a) > 1 and a[1] is not None:
                cur_node.right = Node(a[1])
                a.append(cur_node.right)
            a = a[2:]

    def print_tree(self):
        print(self.postorder(self.root, ""))
    
    def postorder(self, node, traversal):
        # left -> root -> right
        if node:
            traversal = self.postorder(node.left, traversal)
            traversal += f"{node.val}"
            traversal = self.postorder(node.right, traversal)
        return traversal
    
    def find_kth(self, k):
        array = self._find_kth(self.root, [])
        return(array[k-1])

    def _find_kth1(self, node, k, array):
        # left -> root -> right
        if node:
            self._find_kth(node.left, k, array)
            if len(array) == k:
                return
            array.append(node.val)
            self._find_kth(node.right, k, array)
        return array

    def _find_kth2(self, node, array):
        # left -> root -> right
        if node:
            self._find_kth(node.left, array)
            array.append(node.val)
            self._find_kth(node.right, array)
        return array

    def _find_kth(self, k):
        #Initialize stack and initial position
        stack = []
        n = 0

        cur_node = self.root
        while cur_node or stack:
            #Traverse left subtree
            while cur_node:
                stack.append(cur_node)
                cur_node = cur_node.left

            #Visiting current node
            cur_node = stack.pop()
            n += 1
            if n == k:
                return cur_node.val 
            #Traversing the right side

            cur_node = cur_node.right
        return None



    
a = [3, 1, 4, None, 2]

t = BST()
t.create_bst(a)
# t.print_tree()
print(t._find_kth(2))

```
